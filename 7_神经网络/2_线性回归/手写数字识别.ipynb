# 手写数字识别

```python
import tensorflow  as tf
from tensorflow.examples.tutorials.mnist import input_data
import numpy as np
import matplotlib.pyplot as plt
tf.compat.v1.disable_eager_execution()

#导入数据
mnist=input_data.read_data_sets("mnist_data",one_hot=True)

#每个批次的大小
batch_size=500
#计算一共有多少批次
n_batch=mnist.train.num_examples
#定义两个placeholder
x=tf.compat.v1.placeholder(tf.float32,[None,784]) #输入784个神经元
y=tf.compat.v1.placeholder(tf.float32,[None,10])  #输出10个神经元
keep_prob=tf.compat.v1.placeholder(tf.float32)    #输入dropout的百分比

'''创建一层神经网络'''
w1=tf.Variable(tf.ones([784,10]))
b1=tf.Variable(tf.ones([10]))
#激活函数
pre=tf.nn.softmax(tf.matmul(x,w1)+b1)

'''二次代价函数'''
loss=tf.reduce_mean(tf.square(y-pre))
#loss=tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(labels=y,logits=pre))  #交叉熵
'''梯度下降'''
train_step=tf.compat.v1.train.GradientDescentOptimizer(learning_rate=0.7).minimize(loss)
'''创建会话'''
init=tf.compat.v1.global_variables_initializer()
#结果存放在一个布尔值列表中
correct=tf.equal(tf.argmax(y,1),tf.argmax(pre,1))  #argmax返回一维张量中最大的值
#求准确率
accuracy=tf.reduce_mean(tf.cast(correct,tf.float32))
#创建会话
with tf.compat.v1.Session() as sess:
    sess.run(init)  
    for epoch in range(20):
        for batch in range(n_batch):
            batch_x,batch_y=mnist.train.next_batch(batch_size)
            sess.run(train_step,feed_dict={x:batch_x,y:batch_y,keep_prob:1})

        test_acc=sess.run(accuracy,feed_dict={x:mnist.test.images,y:mnist.test.labels,keep_prob:1})
        train_acc=sess.run(accuracy,feed_dict={x:mnist.train.images,y:mnist.train.labels,keep_prob:1})
        print("epoch"+str(epoch),"test="+str(test_acc),"train"+str(train_acc))
```

## 优化
* 创建多层神经元
* dropout

```python
import tensorflow  as tf
from tensorflow.examples.tutorials.mnist import input_data
import numpy as np
import matplotlib.pyplot as plt
tf.compat.v1.disable_eager_execution()

#导入数据
mnist=input_data.read_data_sets("mnist_data",one_hot=True)

#每个批次的大小
batch_size=500
#计算一共有多少批次
n_batch=mnist.train.num_examples
#定义两个placeholder
x=tf.compat.v1.placeholder(tf.float32,[None,784]) #输入784个神经元
y=tf.compat.v1.placeholder(tf.float32,[None,10])  #输出10个神经元
prob=tf.compat.v1.placeholder(tf.float32)    #输入dropout的百分比

'''创建神经网络的输入层'''
w1=tf.Variable(tf.compat.v1.truncated_normal([784,1000],stddev=0.1))
b1=tf.Variable(tf.zeros([1000])+0.1)
#激活函数
l1=tf.nn.tanh(tf.matmul(x,w1)+b1)
#dropout
l1_drop=tf.nn.dropout(l1,rate=prob)
'''构建神经网络的隐藏层'''
#layer1  1000个神经元
w2=tf.Variable(tf.compat.v1.truncated_normal([1000,500],stddev=0.1))
b2=tf.Variable(tf.zeros([500])+0.1)
l2=tf.nn.tanh(tf.matmul(l1_drop,w2)+b2)
l2_drop=tf.nn.dropout(l2,rate=prob)  #每个元素被丢弃的概率

'''输出层'''
#是个神经元
w3=tf.Variable(tf.compat.v1.truncated_normal([500,10],stddev=0.1))
b3=tf.Variable(tf.zeros([10])+0.1)
pre=tf.nn.softmax(tf.matmul(l2_drop,w3)+b3)

'''二次代价函数'''
loss=tf.reduce_mean(tf.square(y-pre))
#loss=tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(labels=y,logits=pre))  #交叉熵
'''梯度下降'''
train_step=tf.compat.v1.train.GradientDescentOptimizer(learning_rate=0.5).minimize(loss)
'''创建会话'''
init=tf.compat.v1.global_variables_initializer()
#结果存放在一个布尔值列表中
correct=tf.equal(tf.argmax(y,1),tf.argmax(pre,1))  #argmax返回一维张量中最大的值
#求准确率
accuracy=tf.reduce_mean(tf.cast(correct,tf.float32))
#创建会话
with tf.compat.v1.Session() as sess:
    sess.run(init)  
    for epoch in range(20):
        for batch in range(n_batch):
            batch_x,batch_y=mnist.train.next_batch(batch_size)
            sess.run(train_step,feed_dict={x:batch_x,y:batch_y,prob:0.2})  #隐藏层进行dropout

        test_acc=sess.run(accuracy,feed_dict={x:mnist.test.images,y:mnist.test.labels,prob:0})
        train_acc=sess.run(accuracy,feed_dict={x:mnist.train.images,y:mnist.train.labels,prob:0})
        print("epoch"+str(epoch),"test="+str(test_acc),"train="+str(train_acc))
```
